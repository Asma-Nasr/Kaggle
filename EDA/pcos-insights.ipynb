```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/kaggle/input/diet-exercise-and-pcos-insights/Cleaned-Data.csv')
```

```python
data.head(3)
```

```python
data.columns
```

```python
# Histogram
data['PCOS'].hist(bins=30)
plt.title('Distribution of Column Name')
plt.xlabel('Column Name')
plt.ylabel('Frequency')
plt.show()
```

```python
# Histogram
data['Insulin_Resistance'].hist(bins=30)
plt.title('Distribution of Column Name')
plt.xlabel('Column Name')
plt.ylabel('Frequency')
plt.show()
```

```python
data['Weight_kg'].hist(bins=20)
plt.title('Weight Distribution')
plt.xlabel('Weight (kg)')
plt.ylabel('Frequency')
plt.show()
```

```python
data['Marital_Status'].value_counts().plot(kind='bar')
plt.title('Marital Status Distribution')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.show()
```

```python
avg_weight = data.groupby(['Menstrual_Irregularity', 'PCOS'])['Weight_kg'].mean().unstack()
avg_weight.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Average Weight by Menstrual_Irregularity and PCOS Status')
plt.xlabel('Menstrual_Irregularity ')
plt.ylabel('Average Weight (kg)')
plt.show()
```

```python
avg_weight = data.groupby(['Marital_Status', 'PCOS'])['Weight_kg'].mean().unstack()
avg_weight.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Average Weight by Marital Status and PCOS Status')
plt.xlabel('Marital Status')
plt.ylabel('Average Weight (kg)')
plt.show()
```
