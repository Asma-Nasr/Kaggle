```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,ConfusionMatrixDisplay,roc_auc_score,classification_report,precision_score,recall_score,f1_score
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.read_csv('/kaggle/input/loan-approval-prediction-dataset/loan_approval_dataset.csv')
data.shape
```

```python
data.head(3)
```

## Missing data?

```python
data.isna().sum()
```

## Number of Numerical & Categorical columns

```python
numerical_cols = data.select_dtypes(include=['number']).columns.tolist()
print("Numerical Columns:", len(numerical_cols))
```

```python
categorical_cols = data.select_dtypes(include=['object']).columns.tolist()
print("Categorical Columns:", len(categorical_cols))
```

```python
for col in numerical_cols:
    plt.figure(figsize=(10, 4))
    sns.histplot(data[col], bins=30, kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
for col in categorical_cols:
    plt.figure(figsize=(8, 4))
    sns.countplot(data=data, x=col)
    plt.title(f'Count of {col}')
    plt.xticks(rotation=45)
    plt.show()
```

## Convert Categorical features to Numerical 

```python
lb = LabelEncoder()
for col in categorical_cols:
    data[col] = lb.fit_transform(data[col])
```

```python
data.head(2)
```

### Data Splitting

```python
X = data.drop(columns=['loan_id', ' loan_status'])
y = data[' loan_status']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

### Random Forest Model

```python
model = RandomForestClassifier()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
```

### Accuracy

```python
print(f'Accuracy: {accuracy_score(y_test,y_pred)*100:.2f}%')
```

### ROC Score

```python
print(f'ROC score: {roc_auc_score(y_test,y_pred)*100:.2f}')
```

### Precision Score

```python
print(f'Precision score:{precision_score(y_test,y_pred):.2f}')
```

### Recall Score

```python
print(f'Recall score: {recall_score(y_test,y_pred):.2f}')
```

### F1-Score

```python
print(f'F1-score: {f1_score(y_test,y_pred):.2f}')
```

### Confusion Matrix

```python
cm = confusion_matrix(y_test,y_pred)
disp = ConfusionMatrixDisplay(cm)
disp.plot(ax=plt.gca())
plt.title('Confusion Matrix')
plt.show()
```

### Classification Report

```python
print('Classification Report\n',classification_report(y_test,y_pred))
```
